#### Using RAG (with helpers)

```python
# load required libraries
from dotenv import load_dotenv
load_dotenv()
from ingest import load_faq_data, build_index
from rag_helper import RAGBase
from openai import OpenAI
```

```python
# load data and build index
documents = load_faq_data()
index = build_index(documents)

# start the OpenAI client
openai_client = OpenAI()

# define the RAG
assistant = RAGBase(
    index=index,
    llm_client=openai_client,
)

# try asking a question to RAG
answer = assistant.rag("I just discovered the course. Can I join now?")
print(answer)
```

```python
# overwrite instructions for a different behaviour
custom_instructions = """
You're a course teaching assistant.
Answer the QUESTION based on the CONTEXT from the FAQ database.
Use only the facts from the CONTEXT when answering the QUESTION.
""".strip()

assistant = RAGBase(
    index=index,
    llm_client=openai_client,
    instructions=custom_instructions,
)

# try asking questions
assistant.rag("How do I get a certificate?")
assistant.rag("Can I still join the course after it started?")
```
